

1. Carga y Preparaci칩n de Datos


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
titanic = pd.read_csv('Titanic.csv')

# Mostrar informaci칩n b치sica
print(titanic.info())
print(titanic.describe())
```

2. Visualizaciones Realizadas
a) Gr치fico de barras: Supervivencia por sexo

```python
sns.countplot(x='Sex', hue='Survived', data=titanic)
plt.title('Supervivencia por sexo')
plt.show()
```

Interpretaci칩n: Este gr치fico muestra que la mayor칤a de las mujeres sobrevivieron (alto conteo en Survived=1 para Sex=female), mientras que la mayor칤a de los hombres no sobrevivieron (alto conteo en Survived=0 para Sex=male), lo que refleja la pol칤tica de "mujeres y ni침os primero" durante el rescate.

b) Distribuci칩n de edades por clase

```python
sns.displot(data=titanic, x='Age', hue='Pclass', kind='kde', fill=True)
plt.title('Distribuci칩n de edades por clase')
plt.show()
```

Interpretaci칩n: La distribuci칩n muestra que los pasajeros de primera clase (Pclass=1) tend칤an a ser mayores que los de tercera clase (Pclass=3), con una mayor concentraci칩n de adultos j칩venes en tercera clase.

c) Boxplot de tarifas por puerto de embarque

```python
sns.boxplot(x='Embarked', y='Fare', data=titanic)
plt.title('Tarifas por puerto de embarque')
plt.show()
```

Interpretaci칩n: Los pasajeros que embarcaron en Cherburgo (C) pagaron tarifas m치s altas en promedio, lo que podr칤a indicar que eran pasajeros de clases m치s altas.

d) Gr치fico de viol칤n: Edades por supervivencia

```python
sns.violinplot(x='Survived', y='Age', data=titanic)
plt.title('Distribuci칩n de edades por supervivencia')
plt.show()
```

Interpretaci칩n: La distribuci칩n de edades es similar entre sobrevivientes y no sobrevivientes, aunque hay una mayor concentraci칩n de ni침os entre los sobrevivientes (Survived=1).

e) Diagrama de dispersi칩n: Tarifas vs. Edades por Supervivencia

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic)
plt.title('Tarifas vs. Edades por Supervivencia')
plt.show()
```

Interpretaci칩n: Los pasajeros que pagaron tarifas m치s altas (probablemente de primera clase) tuvieron mayores tasas de supervivencia, especialmente entre los adultos.

f) Gr치fico de barras apiladas: Supervivencia vs. Clase agrupado por puerto

```python
pd.crosstab([titanic['Embarked'], titanic['Pclass']], titanic['Survived']).plot(kind='bar', stacked=True)
plt.title('Supervivencia vs. Clase por Puerto de Embarque')
plt.show()
```

Interpretaci칩n: Muestra c칩mo la combinaci칩n de clase y puerto de embarque afect칩 las tasas de supervivencia, con mejores resultados para primera clase en todos los puertos.

g) Pairplot de variables seleccionadas

```python
sns.pairplot(titanic[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle('Relaciones entre Age, Fare y Pclass por Supervivencia', y=1.02)
plt.show()
```

Interpretaci칩n: Revela patrones complejos entre las variables, mostrando que la combinaci칩n de edad joven, tarifas altas y primera clase estaba asociada con mayores tasas de supervivencia.

3. Conclusiones
El an치lisis visual revela varios patrones interesantes:

G칠nero: Las mujeres tuvieron tasas de supervivencia significativamente m치s altas que los hombres.

Clase social: Los pasajeros de primera clase, especialmente aquellos que pagaron tarifas m치s altas, ten칤an mayores probabilidades de sobrevivir.

Edad: Los ni침os (especialmente en primera y segunda clase) tuvieron altas tasas de supervivencia.

Puerto de embarque: El puerto de embarque estaba relacionado con la clase social y tarifa, lo que a su vez afectaba la supervivencia.

Estas visualizaciones demuestran el poder de Seaborn para explorar y comunicar patrones en conjuntos de datos complejos de manera efectiva. Los resultados confirman lo que sabemos hist칩ricamente sobre el desastre del Titanic: que las mujeres y ni침os de clases sociales m치s altas tuvieron prioridad en los botes salvavidas.
