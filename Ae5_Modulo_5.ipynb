

1. Carga y Preparación de Datos


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
titanic = pd.read_csv('Titanic.csv')

# Mostrar información básica
print(titanic.info())
print(titanic.describe())
```

2. Visualizaciones Realizadas
a) Gráfico de barras: Supervivencia por sexo

```python
sns.countplot(x='Sex', hue='Survived', data=titanic)
plt.title('Supervivencia por sexo')
plt.show()
```

Interpretación: Este gráfico muestra que la mayoría de las mujeres sobrevivieron (alto conteo en Survived=1 para Sex=female), mientras que la mayoría de los hombres no sobrevivieron (alto conteo en Survived=0 para Sex=male), lo que refleja la política de "mujeres y niños primero" durante el rescate.

b) Distribución de edades por clase

```python
sns.displot(data=titanic, x='Age', hue='Pclass', kind='kde', fill=True)
plt.title('Distribución de edades por clase')
plt.show()
```

Interpretación: La distribución muestra que los pasajeros de primera clase (Pclass=1) tendían a ser mayores que los de tercera clase (Pclass=3), con una mayor concentración de adultos jóvenes en tercera clase.

c) Boxplot de tarifas por puerto de embarque

```python
sns.boxplot(x='Embarked', y='Fare', data=titanic)
plt.title('Tarifas por puerto de embarque')
plt.show()
```

Interpretación: Los pasajeros que embarcaron en Cherburgo (C) pagaron tarifas más altas en promedio, lo que podría indicar que eran pasajeros de clases más altas.

d) Gráfico de violín: Edades por supervivencia

```python
sns.violinplot(x='Survived', y='Age', data=titanic)
plt.title('Distribución de edades por supervivencia')
plt.show()
```

Interpretación: La distribución de edades es similar entre sobrevivientes y no sobrevivientes, aunque hay una mayor concentración de niños entre los sobrevivientes (Survived=1).

e) Diagrama de dispersión: Tarifas vs. Edades por Supervivencia

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic)
plt.title('Tarifas vs. Edades por Supervivencia')
plt.show()
```

Interpretación: Los pasajeros que pagaron tarifas más altas (probablemente de primera clase) tuvieron mayores tasas de supervivencia, especialmente entre los adultos.

f) Gráfico de barras apiladas: Supervivencia vs. Clase agrupado por puerto

```python
pd.crosstab([titanic['Embarked'], titanic['Pclass']], titanic['Survived']).plot(kind='bar', stacked=True)
plt.title('Supervivencia vs. Clase por Puerto de Embarque')
plt.show()
```

Interpretación: Muestra cómo la combinación de clase y puerto de embarque afectó las tasas de supervivencia, con mejores resultados para primera clase en todos los puertos.

g) Pairplot de variables seleccionadas

```python
sns.pairplot(titanic[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle('Relaciones entre Age, Fare y Pclass por Supervivencia', y=1.02)
plt.show()
```

Interpretación: Revela patrones complejos entre las variables, mostrando que la combinación de edad joven, tarifas altas y primera clase estaba asociada con mayores tasas de supervivencia.

3. Conclusiones
El análisis visual revela varios patrones interesantes:

Género: Las mujeres tuvieron tasas de supervivencia significativamente más altas que los hombres.

Clase social: Los pasajeros de primera clase, especialmente aquellos que pagaron tarifas más altas, tenían mayores probabilidades de sobrevivir.

Edad: Los niños (especialmente en primera y segunda clase) tuvieron altas tasas de supervivencia.

Puerto de embarque: El puerto de embarque estaba relacionado con la clase social y tarifa, lo que a su vez afectaba la supervivencia.

Estas visualizaciones demuestran el poder de Seaborn para explorar y comunicar patrones en conjuntos de datos complejos de manera efectiva. Los resultados confirman lo que sabemos históricamente sobre el desastre del Titanic: que las mujeres y niños de clases sociales más altas tuvieron prioridad en los botes salvavidas.
